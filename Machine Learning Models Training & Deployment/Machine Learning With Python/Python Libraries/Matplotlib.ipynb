```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,6,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
plt.show()
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Outside')
axes2.plot(y,x)
axes2.set_title('Inside')
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.show()
```

```python
axes
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
plt.show()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(4,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cube')
ax.legend(loc=0)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=6)
plt.show()
```
